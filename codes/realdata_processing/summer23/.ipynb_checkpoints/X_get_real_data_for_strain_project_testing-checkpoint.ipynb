```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
```

```python
realdata_dir = '/home/sdybing/gnss-picker/data/realdata/summer23/'
```

```python
real_data = h5py.File(realdata_dir + 'demean_realdata_rembad.hdf5', 'r')
real_data = real_data['demean_realdata_rembad'][:,:]
```

```python
metadata = np.load(realdata_dir + 'real_metadata_rembad_w_gauss_pos_mag.npy')

# Columns:
# 0: station name
# 1: date sample came from
# 2: sample start time
# 3: sample end time
# 4: random n counter
# 5: sample P-wave arrives at (gauss pos)
# 6: earthquake magnitude
```

```python
j = np.where((metadata[:,6] == '7.10') & (metadata[:,0] == 'P595') | (metadata[:,6] == '6.40') & (metadata[:,0] == 'P595'))[0]
print(j)
```

```python
subset_metadata = metadata[j]
subset_metadata.shape
```

```python
subset_real_data = real_data[j]
```

```python
subset_real_data.shape
```

```python
subset_metadata[1]
```

```python
plt.subplots(figsize = (10,4))

plt.subplot(131)
plt.plot(subset_real_data[1,0:128], label = 'RT-HR GNSS')

peak_gnss_N = subset_real_data[1,0:128].copy()
for k in range(0,len(subset_real_data[1,0:128])):
    if k == 0:
        gnss = subset_real_data[1,0:128][0]
        max_gnss = gnss
    else:
        gnss = subset_real_data[1,0:128][:k+1]
        max_gnss = max(gnss)
    peak_gnss_N[k] = max_gnss 

plt.plot(peak_gnss_N, color = 'red', label = 'Peak GNSS')
plt.axvline(float(subset_metadata[1,5]), color = 'orange', label = 'P-wave arrival')
plt.xlim(54,128)
plt.legend()

plt.subplot(132)
plt.plot(subset_real_data[1,128:256])

peak_gnss_E = subset_real_data[1,128:256].copy()
for k in range(0,len(subset_real_data[1,128:256])):
    if k == 0:
        gnss = subset_real_data[1,128:256][0]
        max_gnss = gnss
    else:
        gnss = subset_real_data[1,128:256][:k+1]
        max_gnss = max(gnss)
    peak_gnss_E[k] = max_gnss 

plt.plot(peak_gnss_E, color = 'red')
plt.axvline(float(subset_metadata[1,5]), color = 'orange')
plt.xlim(54,128)

plt.subplot(133)
plt.plot(subset_real_data[1,256:])

peak_gnss_Z = subset_real_data[1,256:].copy()
for k in range(0,len(subset_real_data[1,256:])):
    if k == 0:
        gnss = subset_real_data[1,256:][0]
        max_gnss = gnss
    else:
        gnss = subset_real_data[1,256:][:k+1]
        max_gnss = max(gnss)
    peak_gnss_Z[k] = max_gnss

plt.plot(peak_gnss_Z, color = 'red')
plt.axvline(float(subset_metadata[1,5]), color = 'orange')
plt.xlim(54,128)
plt.show();




plt.subplots(figsize = (10,4))
plt.suptitle('Log y axis peak GNSS')

plt.subplot(131)
plt.plot(peak_gnss_N, color = 'red', label = 'Peak GNSS')
plt.axvline(float(subset_metadata[1,5]), color = 'orange', label = 'P-wave arrival')
plt.xlim(54,128)
plt.semilogy()
plt.legend()

plt.subplot(132)
plt.plot(peak_gnss_E, color = 'red')
plt.axvline(float(subset_metadata[1,5]), color = 'orange')
plt.xlim(54,128)
plt.semilogy()

plt.subplot(133)
plt.plot(peak_gnss_Z, color = 'red')
plt.axvline(float(subset_metadata[1,5]), color = 'orange')
plt.xlim(54,128)
plt.semilogy()

plt.show();

```
