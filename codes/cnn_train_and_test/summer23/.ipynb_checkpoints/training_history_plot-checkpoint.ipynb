```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
history = np.load('/home/sdybing/gnss-picker/cnn_models_outputs/newfault_fq_train/models/traindate_2024-10-01/training_history.npy', allow_pickle = 'TRUE').item()

fig = plt.subplots(nrows = 2, ncols = 1, figsize = (10,4), dpi = 400)

plt.subplot(121)
plt.text(x = -18, y = 0.11875, s = '(a)', fontsize = 22)
plt.plot(history['loss'], label = 'Training loss')
plt.plot(history['val_loss'], label = 'Validation loss') 
plt.legend(loc = 'upper right', fontsize = 12)
plt.xlabel('Epoch', fontsize = 12)
plt.ylabel('Loss', fontsize = 12)
plt.tick_params(labelsize = 11)

plt.subplot(122)
plt.text(x = -18, y = 0.0288, s = '(b)', fontsize = 22)
plt.plot(history['accuracy'], label = 'Training accuracy') 
plt.plot(history['val_accuracy'], label = 'Validation accuracy') 
plt.legend(loc = 'lower right', fontsize = 12)
plt.xlabel('Epoch', fontsize = 12)
plt.ylabel('Accuracy', fontsize = 12)
plt.tick_params(labelsize = 11)

plt.subplots_adjust(wspace = 0.25)

# plt.show()

plt.savefig('/home/sdybing/gnss-picker/manuscript_figures/training_history.jpg', format = 'JPG')
plt.close()
```
