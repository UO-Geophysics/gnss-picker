```python
import numpy as np
import matplotlib.pyplot as plt

project = 'july6'
traindate = '2023-07-27'
testdate = '2023-07-28'
traindate_path = '/home/sdybing/gnss-picker/cnn_models_outputs/' + project + '_fq_train/models/traindate_' + traindate + '/'
test_outputs_path = traindate_path + 'data/'
figure_save_dir = traindate_path + 'figures/'
best_thresh = 0.13 # From code 2

results = np.load(test_outputs_path + 'fqtest_metadata_with_results_pgd_snr_locs_hypdists.npy')
# Columns:

# 0: FQ rupture name
# 1: station name
# 2: magnitude
# 3: result (true pos, etc.)
# 4: PGD
# 5: SNR_N
# 6: SNR_E
# 7: SNR_Z
# 8: FQ rupture hypocenter lat
# 9: FQ rupture hypocenter lon
# 10: FQ rupture hypocenter depth
# 11: station lat
# 12: station lon
# 13: hypocentral distance (m)
```

```python
dists = []

for idx in range(len(results)):
    dist = results[idx,13]
    if dist != 'nan':
        dists.append(float(dist))

print(len(dists))
print(min(dists))
print(max(dists))
```

```python
dists_23 = []
dists_34 = []
dists_45 = []
dists_56 = []
dists_67 = []
dists_78 = []
dists_89 = []
dists_90 = []

true_pos = []
false_neg = []

for i in range(len(results)):
    
    if results[i][0] == 'nan':
        pass
    
    else:
        dist = float(results[i][13])
        result = results[i][3]
        
        if dist <= 23000:
            dists_23.append(i)
        elif dist > 23000 and dist <= 24000:
            dists_34.append(i)
        elif dist > 24000 and dist <= 25000:
            dists_45.append(i)            
        elif dist > 25000 and dist <= 26000:
            dists_56.append(i)        
        elif dist > 26000 and dist <= 27000:
            dists_67.append(i)        
        elif dist > 27000 and dist <= 28000:
            dists_78.append(i)        
        elif dist > 28000 and dist <= 29000:
            dists_89.append(i)            
        elif dist > 29000 and dist <= 30000:
            dists_90.append(i)                      
        
        if result == 'true pos': # predicted 1, target 1
            true_pos.append(1)
        
        elif result == 'false neg': # predicted 0, target 1
            false_neg.append(1)
            
        else:
            print('Error at i ' + str(i))
```

```python
number_correct = len(true_pos)
number_incorrect = len(false_neg)
total_number = number_correct + number_incorrect

print('True positive: ' + str(number_correct))
print('False negative: ' + str(number_incorrect))
print('Total number: ' + str(total_number))
print('Overall accuracy: ' + str(100*(np.round(number_correct/total_number,3))) + '%')
```

```python
print(len(dists_23)+len(dists_34)+len(dists_45)+len(dists_56)+len(dists_67)+len(dists_78)+len(dists_89)+len(dists_90))
```
