```python
import numpy as np
import matplotlib.pyplot as plt

project = 'july6'
traindate = '2023-08-26'
testdate = '2023-08-27'
traindate_path = '/home/sdybing/gnss-picker/cnn_models_outputs/' + project + '_fq_train/models/traindate_' + traindate + '/'
test_outputs_path = traindate_path + 'data/'
figure_save_dir = traindate_path + 'figures/'
best_thresh = 0.13 # From code 2

results = np.load(test_outputs_path + 'fqtest_metadata_with_results_pgd_snr.npy')
# Columns:

# 0: FQ rupture name
# 1: station name
# 2: magnitude
# 3: result (true pos, etc.)
# 4: PGD
# 5: SNR_N
# 6: SNR_E
# 7: SNR_Z
```

```python
pgds = []

for idx in range(len(results)):
    pgd = results[idx,4]
    if pgd != 'nan':
        pgds.append(float(pgd))

print(len(pgds))
print(min(np.log10(pgds)))
print(max(np.log10(pgds)))
```

```python
logpgds_2175 = []
logpgds_17515 = []
logpgds_15125 = []
logpgds_1251 = []
logpgds_1075 = []
logpgds_07505 = []
logpgds_05025 = []
logpgds_0250 = []
logpgds_0025 = []

true_pos = []
false_neg = []

for i in range(len(results)):
    
    if results[i][0] == 'nan':
        pass
    
    else:
        logpgd = np.log10(float(results[i][4]))
        result = results[i][3]
        
        if logpgd <= -1.75:
            logpgds_2175.append(i)
        elif logpgd > -1.75 and logpgd <= -1.5:
            logpgds_17515.append(i)            
        elif logpgd > -1.5 and logpgd <= -1.25:
            logpgds_15125.append(i)        
        elif logpgd > -1.25 and logpgd <= -1:
            logpgds_1251.append(i)        
        elif logpgd > -1 and logpgd <= -0.75:
            logpgds_1075.append(i)        
        elif logpgd > -0.75 and logpgd <= -0.5:
            logpgds_07505.append(i)            
        elif logpgd > -0.5 and logpgd <= -0.25:
            logpgds_05025.append(i)        
        elif logpgd > -0.25 and logpgd <= 0:
            logpgds_0250.append(i)        
        elif logpgd > 0 and logpgd <= 0.25:
            logpgds_0025.append(i)                 
        
        if result == 'true pos': # predicted 1, target 1
            true_pos.append(1)
        
        elif result == 'false neg': # predicted 0, target 1
            false_neg.append(1)
            
        else:
            print('Error at i ' + str(i))
```

```python
number_correct = len(true_pos)
number_incorrect = len(false_neg)
total_number = number_correct + number_incorrect

print('True positive: ' + str(number_correct))
print('False negative: ' + str(number_incorrect))
print('Total number: ' + str(total_number))
print('Overall accuracy: ' + str(100*(np.round(number_correct/total_number,3))) + '%')
```
