```python
import numpy as np
import matplotlib.pyplot as plt

project = 'newfault'
traindate = '2024-10-01'
testdate = '2024-10-02'
traindate_path = '/home/sdybing/gnss-picker/cnn_models_outputs/' + project + '_fq_train/models/traindate_' + traindate + '/'
test_outputs_path = traindate_path + 'data/'
figure_save_dir = traindate_path + 'figures/'
best_thresh = 0.135 # From code 2

SNREs = np.load(test_outputs_path + 'fakequakes_testing/fqtest_data_SNRs_E.npy')
results = np.load(test_outputs_path + 'fakequakes_testing/fqtest_metadata_with_results_pgd_snr.npy')
# Columns:

# 0: FQ rupture name
# 1: station name
# 2: magnitude
# 3: result (true pos, etc.)
# 4: PGD
# 5: SNR_N
# 6: SNR_E
# 7: SNR_Z
```

```python
rangeSNRs = []

for i in range(len(SNREs)):
    if float(SNREs[i]) == 0 or SNREs[i] == 'nan':
        pass
    else:
        rangeSNRs.append(float(SNREs[i]))

print(min(np.log10(rangeSNRs)))
print(max(np.log10(rangeSNRs)))
```

```python
snr_252 = []
snr_215 = []
snr_151 = []
snr_105 = []
snr_050 = []
snr_005 = []
snr_051 = []
snr_115 = []
snr_152 = []
snr_225 = []
snr_253 = []

true_pos = []
false_neg = []

for i in range(len(SNREs)):
    
    logSNRE = np.log10(float(SNREs[i]))
    result = results[i][3]
    
    if SNREs[i] == 'nan' or float(SNREs[i]) == 0:
        pass
    
    else:
        if logSNRE <= -2:
            snr_252.append(i)
        elif logSNRE > -2 and logSNRE <= -1.5:
            snr_215.append(i)
        elif logSNRE > -1.5 and logSNRE <= -1:
            snr_151.append(i)       
        elif logSNRE > -1 and logSNRE <= -0.5:
            snr_105.append(i)        
        elif logSNRE > -0.5 and logSNRE <= 0:
            snr_050.append(i)            
        elif logSNRE > 0 and logSNRE <= 0.5:
            snr_005.append(i)            
        elif logSNRE > 0.5 and logSNRE <= 1:
            snr_051.append(i)            
        elif logSNRE > 1 and logSNRE <= 1.5:
            snr_115.append(i)            
        elif logSNRE > 1.5 and logSNRE <= 2:
            snr_152.append(i)            
        elif logSNRE > 2 and logSNRE <= 2.5:
            snr_225.append(i)
        elif logSNRE > 2.5 and logSNRE <= 3:
            snr_253.append(i)
            
        if result == 'true pos': # predicted 1, target 1
            true_pos.append(1)
        
        elif result == 'false neg': # predicted 0, target 1
            false_neg.append(1)
        
        else:
            print('Error at i ' + str(i))
```

```python
number_correct = len(true_pos)
number_incorrect = len(false_neg)
total_number = number_correct + number_incorrect

print('True positive: ' + str(number_correct))
print('False negative: ' + str(number_incorrect))
print('Total number: ' + str(total_number))
print('Overall accuracy: ' + str(100*(np.round(number_correct/total_number,2))) + '%')
```
