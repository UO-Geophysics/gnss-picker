```python
mean_snrs = []
only_good_ones = []

for idx in range(len(results)):
    
    snrn = results[idx,5]
    snre = results[idx,6]
    snrz = results[idx,7]
    
    if snrn == 'nan' or float(snrn) == 0:
        mean_snrs.append('nan')
    
    else:
        mean_snr = (float(snrn) + float(snre) + float(snrz))/3
        mean_snrs.append(mean_snr)
        only_good_ones.append(mean_snr)
```
