```python
import numpy as np
import matplotlib.pyplot as plt

project = 'newfault'
traindate = '2024-10-01'
testdate = '2024-10-02'
traindate_path = '/home/sdybing/gnss-picker/cnn_models_outputs/' + project + '_fq_train/models/traindate_' + traindate + '/'
test_outputs_path = traindate_path + 'data/'
figure_save_dir = traindate_path + 'figures/'
best_thresh = 0.135 # From code 2

SNRZs = np.load(test_outputs_path + 'fakequakes_testing/fqtest_data_SNRs_Z.npy')
results = np.load(test_outputs_path + 'fakequakes_testing/fqtest_metadata_with_results_pgd_snr.npy')
# Columns:

# 0: FQ rupture name
# 1: station name
# 2: magnitude
# 3: result (true pos, etc.)
# 4: PGD
# 5: SNR_N
# 6: SNR_E
# 7: SNR_Z
```

```python
rangeSNRs = []

for i in range(len(SNRZs)):
    if float(SNRZs[i]) == 0 or SNRZs[i] == 'nan':
        pass
    else:
        rangeSNRs.append(float(SNRZs[i]))

print(min(np.log10(rangeSNRs)))
print(max(np.log10(rangeSNRs)))
```

```python
snr_655 = []
snr_555 = []
snr_545 = []
snr_454 = []
snr_435 = []
snr_353 = []
snr_325 = []
snr_252 = []
snr_215 = []
snr_151 = []
snr_105 = []
snr_050 = []
snr_005 = []
snr_051 = []
snr_115 = []

true_pos = []
false_neg = []

for i in range(len(SNRZs)):
    
    logSNRZ = np.log10(float(SNRZs[i]))
    result = results[i][3]
    
    if SNRZs[i] == 'nan' or float(SNRZs[i]) == 0:
        pass
    
    else:
        if logSNRZ <= -5.5:
            snr_655.append(i)
        elif logSNRZ > -5.5 and logSNRZ <= -5:
            snr_555.append(i)
        elif logSNRZ > -5 and logSNRZ <= -4.5:
            snr_545.append(i)
        elif logSNRZ > -4.5 and logSNRZ <= -4:
            snr_454.append(i)
        elif logSNRZ > -4 and logSNRZ <= -3.5:
            snr_435.append(i)
        elif logSNRZ > -3.5 and logSNRZ <= -3:
            snr_353.append(i)
        elif logSNRZ > -3 and logSNRZ <= -2.5:
            snr_325.append(i)
        elif logSNRZ > -2.5 and logSNRZ <= -2:
            snr_252.append(i)
        elif logSNRZ > -2 and logSNRZ <= -1.5:
            snr_215.append(i)
        elif logSNRZ > -1.5 and logSNRZ <= -1:
            snr_151.append(i)       
        elif logSNRZ > -1 and logSNRZ <= -0.5:
            snr_105.append(i)        
        elif logSNRZ > -0.5 and logSNRZ <= 0:
            snr_050.append(i)            
        elif logSNRZ > 0 and logSNRZ <= 0.5:
            snr_005.append(i)            
        elif logSNRZ > 0.5 and logSNRZ <= 1:
            snr_051.append(i)            
        elif logSNRZ > 1 and logSNRZ <= 1.5:
            snr_115.append(i)            
        elif logSNRZ > 1.5 and logSNRZ <= 2:
            snr_152.append(i)            
        elif logSNRZ > 2 and logSNRZ <= 2.5:
            snr_225.append(i)
        else:
            print('Error at i ' + str(i))
            
        if result == 'true pos': # predicted 1, target 1
            true_pos.append(1)
        
        elif result == 'false neg': # predicted 0, target 1
            false_neg.append(1)
        
        else:
            print('Error at i ' + str(i))
```

```python
number_correct = len(true_pos)
number_incorrect = len(false_neg)
total_number = number_correct + number_incorrect

print('True positive: ' + str(number_correct))
print('False negative: ' + str(number_incorrect))
print('Total number: ' + str(total_number))
print('Overall accuracy: ' + str(100*(np.round(number_correct/total_number,2))) + '%')
```
