```python
import numpy as np
import matplotlib.pyplot as plt

project = 'july6'
traindate = '2023-07-27'
testdate = '2023-07-28'
traindate_path = '/home/sdybing/gnss-picker/cnn_models_outputs/' + project + '_fq_train/models/traindate_' + traindate + '/'
test_outputs_path = traindate_path + 'data/'
figure_save_dir = traindate_path + 'figures/'
best_thresh = 0.13 # From code 2

results = np.load(test_outputs_path + 'fqtest_metadata_with_results_pgd_snr_locs_hypdists.npy')
# Columns:

# 0: FQ rupture name
# 1: station name
# 2: magnitude
# 3: result (true pos, etc.)
# 4: PGD
# 5: SNR_N
# 6: SNR_E
# 7: SNR_Z
# 8: FQ rupture hypocenter lat
# 9: FQ rupture hypocenter lon
# 10: FQ rupture hypocenter depth
# 11: station lat
# 12: station lon
# 13: hypocentral distance (m)
```

```python
dists = []

for idx in range(len(results)):
    dist = results[idx,13]
    if dist != 'nan':
        dists.append(float(dist))

print(len(dists))
print(min(dists))
print(max(dists))
```

```python
dists_04 = []
dists_48 = []
dists_812 = []
dists_1216 = []
dists_1620 = []
dists_2024 = []
dists_2428 = []
dists_2832 = []
dists_3236 = []

true_pos = []
false_neg = []

for i in range(len(results)):
    
    if results[i][0] == 'nan':
        pass
    
    else:
        dist = float(results[i][13])/1000
        result = results[i][3]
        
        if dist <= 40:
            dists_04.append(i)
        elif dist > 40 and dist <= 80:
            dists_48.append(i)
        elif dist > 80 and dist <= 120:
            dists_812.append(i)            
        elif dist > 120 and dist <= 160:
            dists_1216.append(i)        
        elif dist > 160 and dist <= 200:
            dists_1620.append(i)        
        elif dist > 200 and dist <= 240:
            dists_2024.append(i)        
        elif dist > 240 and dist <= 280:
            dists_2428.append(i)            
        elif dist > 280 and dist <= 320:
            dists_2832.append(i) 
        elif dist > 320 and dist <= 360:
            dists_3236.append(i)
        
        if result == 'true pos': # predicted 1, target 1
            true_pos.append(1)
        
        elif result == 'false neg': # predicted 0, target 1
            false_neg.append(1)
            
        else:
            print('Error at i ' + str(i))
```

```python
number_correct = len(true_pos)
number_incorrect = len(false_neg)
total_number = number_correct + number_incorrect

print('True positive: ' + str(number_correct))
print('False negative: ' + str(number_incorrect))
print('Total number: ' + str(total_number))
print('Overall accuracy: ' + str(100*(np.round(number_correct/total_number,3))) + '%')
```
