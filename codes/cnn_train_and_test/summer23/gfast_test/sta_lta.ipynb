```python
# Kawamoto STA/LTA values were 60 and 600
sta = 5
lta = 30

for idx in range(len(data)):
    
    waveforms = data[idx]

    n = waveforms[:,0]
    e = waveforms[:,1]
    z = waveforms[:,2]
    
    # Run picker on each channel and each trace
         
    # Pick on each trace individually
    npick = np.zeros((n.shape))
    epick = np.zeros((e.shape))
    zpick = np.zeros((z.shape))
    
    npick = classic_sta_lta(n, sta, lta)
    epick = classic_sta_lta(e, sta, lta)
    zpick = classic_sta_lta(z, sta, lta)
    
#     for k in range(len(npick)):
#         npick[k] = picker(n[k], sta, lta)
#         epick[k] = picker(e[k], sta, lta)
#         zpick[k] = picker(z[k], sta, lta)
    
#     print(npick)
    
    plt.figure(figsize = (10,2), dpi = 100)
    
    ax1 = plt.subplot(131)
    ax1.plot(n, color = 'C0')
    ax2 = ax1.twinx()
    ax2.plot(npick, color = 'black', linestyle = '--')
    for i in range(len(npick)):
        if npick[i] >= 1.5:
            plt.axvline(x = i, color = 'red')
    
    ax3 = plt.subplot(132)
    ax3.plot(e, color = 'C1')
    ax4 = ax3.twinx()
    ax4.plot(epick, color = 'red', linestyle = '--')
    ax4.plot(targets[idx], color = 'black', linestyle = '--')
    
    ax5 = plt.subplot(133)
    ax5.plot(z, color = 'C2')
    ax6 = ax5.twinx()
    ax6.plot(zpick, color = 'black', linestyle = '--')
    
    plt.show()
```
